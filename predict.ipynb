```python
BTC['y'] = BTC['Close']
BTC['ds'] = BTC.index
BTC.head()
```

```python
m = Prophet()
m.fit(BTC)
future = m.make_future_dataframe(periods=365)
forecast=m.predict(future)
```

```python
forecast.tail()
```

```python
fig = m.plot(forecast)
```

```python
plot_components_plotly(m, forecast)
```

```python
fig = plot_plotly(m,forecast)
fig
```

```python
fig = m.plot(forecast)
a = add_changepoints_to_plot(fig.gca(),m,forecast)
```

```python
m = Prophet(changepoint_prior_scale=0.005)
forecast = m.fit(BTC).predict(future)
fig=m.plot(forecast)
```
